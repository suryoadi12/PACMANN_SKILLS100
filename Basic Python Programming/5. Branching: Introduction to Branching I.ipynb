Kondisi di atas jika dituliskan pada pernyataan `if` akan sebagai berikut:

```python
# Declare variabel kondisi cuaca
kondisi_cuaca = "Cerah"

# If jika kondisi cuaca cerah
if kondisi_cuaca == "Cerah":
  # Statement jika memenuhi if kondisi jika cuaca cerah
  print("Aku pergi ke pantai")
```

**Contoh lain**: Supermarket A memberikan harga diskon 15% kepada pelanggan yang bertransaksi belanja lebih dari Rp100.000. Wawan tertarik dengan promosi tersebut maka Wawan membeli kebutuhan Wawan dan ternyata didapatkan hasil:

Kondisi di atas dapat di dituliskan ke dalam pernyataan `if`, seperti berikut:

```python
total_belanja_wawan = 110_000
min_belanja = 100_000

# If kondisi jika total belanja wawan lebih besar sama dengan min. belanja
if total_belanja_wawan >= 100_000:
  # Statement yang dikeluarkan jika memenuhi kondisi If diatas
  print("Selamat, anda mendapatkan diskon 15%")

# Menghitung total akhir belanja setelah diskon
total_akhir = total_belanja_wawan-(total_belanja_wawan*0.15)
print("Total akhir belanja anda sebesar", total_akhir)
```

Ternyata total belanja Wawan melebihi syarat ketentuan diskon, maka dia berhak mendapatkan diskon.

Selain menggunakan tipe data text atau angka, kita juga bisa melakukan branching dengan tipe data boolean!



*   Semisal, Tono ingin mengecek apakah produk Ayam Geprek nya bisa mendapatkan promo atau tidak. Maka, dia membuat branching seperti di bawah ini:



```python
# Declare variabel apakah bisa mendapatkan promo
is_promo = True

# If bisa mendapatkan promo
if is_promo == True:
  # Statement yang dikeluarkan jika memenuhi kondisi If diatas
  print("Selamat, anda mendapatkan promo")
```

Misalkan, kamu pergi ke Pantai jika cuaca cerah, namun hari ini hujan maka kamu pergi ke Bioskop.

Maka, jika ditulis bentuk pernyataan **if/else** menjadi:

```python
kondisi_cuaca = "Hujan"

# If kondisi jika cuaca sama dengan cerah
if kondisi_cuaca == "Cerah":
  # Statement yang dikeluarkan jika memenuhi kondisi If diatas
  print("Aku pergi ke pantai")
# Jika tidak memenuhi kondisi If diatas
else:
  print("Aku pergi ke bioskop")
```

Contoh lain: Wawan ingin membeli laptop seharga 20.000.000. Namun, uang yang dimiliki Wawan sebanyak 15.000.000.

```python
harga_laptop = 20_000_000
uang_wawan = 15_000_000

# If uang wawan lebih dari sama dengan harga laptop
if uang_wawan >= harga_laptop:
  # Statement yang dikeluarkan jika memenuhi kondisi If diatas
  print("Wawan dapat membeli laptop")
# Jika tidak memenuhi kondisi If diatas
else:
  print("Wawan harus menabung lagi")
```
