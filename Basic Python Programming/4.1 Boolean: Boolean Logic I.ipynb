**CONTOH CASE**

Seorang data analyst ingin mendata pengguna aplikasi e-commerce apakah pengguna aplikasi setidaknya menggunakan salah satu promo pertama dan promo kedua, maka dapat ditulis sebagai berikut:

```python
#Pengguna menggunakan promo pertama atau promo kedua
is_get_promo_pertama = True
is_get_promo_kedua = True

is_get_promo_pertama or is_get_promo_kedua
```

```python
#Pengguna menggunakan promo pertama saja
is_get_promo_pertama = True
is_get_promo_kedua = False

is_get_promo_pertama or is_get_promo_kedua
```

```python
#Pengguna menggunakan promo kedua saja
is_get_promo_pertama = False
is_get_promo_kedua = True

is_get_promo_pertama or is_get_promo_kedua
```

```python
#Pengguna tidak menggunakan promo
is_get_promo_pertama = False
is_get_promo_kedua = False

is_get_promo_pertama or is_get_promo_kedua
```
