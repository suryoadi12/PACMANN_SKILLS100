**NESTED BRANCHING**



*   Nested branching atau **nested-if** adalah suatu kondisi dimana suatu pernyataan bercabang memiliki **pernyataan bercabang didalamnya**.

*   Biasanya digunakan ketika permasalahan semakin kompleks.

*   Namun, hal ini sebenarnya tidak direkomendasikan.
*   Karena jika nested if sangat panjang, akan mempersulit orang untuk membaca kode.


*   Sehingga akan cukup rawan untuk error.



**CONTOH**

Yana saat ini bekerja sebagai sales sebuah brand sepeda motor. Tiap bulan, Yana perlu untuk menjual 10 sepeda motor. Namun, dari kantor dia bisa mendapatkan insentif jika mampu menjual 20 sepeda motor.

Bulan ini, Yana mampu menjual 30 sepeda motor:

Cek status sales Yana apakah sudah memenuhi target atau belum. Dan jika iya, apakah Yana bisa mendapatkan insentif bulan ini.

Maka flowchart atau alur programnya dapat di gambarkan sebagai berikut:

asdfasdf.JPG

```python
# Declare variabel

sales_yana = 30
target_sales = 10
target_insentif = 20

# If kondisi apakah penjualan Yana lebih besar sama dengan target penjualan
if sales_yana >= target_sales:
  # Nested branching if kondisi apakah penjualan Yana selain mendapatkan target apakah Yana mendapatkan insentif
  if sales_yana >= target_insentif:
    # Statement jika bernilai true (memenuhi target sales dan insentif)
    print("Memenuhi target penjualan dan mendapatkan insentif")
  else:
    # Statement jika bernilai false (tidak memenuhi target insentif tetapi memenuhi target sales)
    print("Memenuhi target penjualan")
else:
  # Statement jika tidak memenuhi target penjualan dan insentif
  print("Tidak memenuhi target penjualan")
```

Dokter merawat seorang pasien yang sedang sakit, dokter memiliki beberapa kondisi diagnosa yaitu:



*   Jika pasien flu tanpa gejala sakit tenggorokan bahwa pasien sakit flu biasa.

*   Jika pasien flu ditambah dengan sakit tenggorokan maka pasien sakit radang tenggorokan.
*   Jika pasien flu ditambah sakit tenggorokan ditambah mual maka pasien sakit tonsilitis.

Ternyata terdapat pasien yang mengidap radang tenggorokan. Ubahlah variabel kondisi_1 dan kondisi_2 agar mengeluarkan diagnosa "pasien sakit radang tenggorokan".

`Hint: Gunakan Boolean Logic`

```python
# Tuliskan jawaban anda di bawah ini

kondisi_1 = True
kondisi_2 = False

if kondisi_1 == True:
  if kondisi_2 == True:
    print("Pasien mengidap sakit tonsilitis")
  else:
    print("Pasien mengidap sakit radang tenggorokan")
else:
  print("Pasien mengidap sakit flu biasa")
```

Buatlah nested if untuk kondisi berikut jika nilai a kurang dari 10, maka lihat apakah Angka tersebut genap atau ganjil, jika genap maka print "angka genap kecil" jika ganjil "angka ganjil kecil" Selain itu print "angka besar".

```python
nilai_a = 5

if nilai_a < 10:
  if nilai_a % 2 == 0:
    print("Angka genap kecil")
  else:
    print("Angka ganjil kecil")
else:
  print("Angka besar")
```
