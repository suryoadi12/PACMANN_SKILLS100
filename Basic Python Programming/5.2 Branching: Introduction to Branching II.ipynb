**ELSE IF/ELIF**

Elif diperlukan ketika branching memiliki lebih dari 2 kondisi.

Misalkan, kamu pergi ke Pantai jika cuaca cerah, jika cuaca hujan maka pergi ke Bioskop, jika tidak memenuhi cuaca keduanya kamu istirahat di Rumah, ternyata cuaca hari ini bersalju.

Maka ilustrasinya seperti berikut:

adsfhlasdjlf.JPG

```python
cuaca_hari_ini = "Badai"

#If kondisi jika cuaca cerah
if cuaca_hari_ini == "Cerah":
  # Statement pertama
  print("Aku pergi ke pantai")
# If kondisi jika cuaca hujan
elif cuaca_hari_ini == "Hujan":
  # Statement kedua
  print("Aku pergi ke bioskop")
# Else jika tidak memenuhi dua kondisi diatas
else:
  # Statement ketiga
  print("Aku istirahat di rumah")
```

**Contoh lain**: Karena didesak kebutuhan untuk menggunakan laptop tersebut, akhirnya Wawan mencari alternatif laptop lain dengan harga yang lebih murah. Setelah melakukan pencarian, dia mendapatkan tiga alternatif, yaitu laptop brand 'Aser' dengan harga 15 juta dan laptop 'Asas' dengan harga 10 juta, dan laptop 'Oman' dengan harga 9 juta.

```python
# Declare variabel
budget_wawan = 10_000_000
laptop_aser = 15_000_000
laptop_asas = 10_000_000
laptop_oman = 9_000_000

# Branching
if budget_wawan >= laptop_aser:
  print("Beli laptop Aser")
elif budget_wawan >= laptop_asas:
  print("Beli laptop Asas")
elif budget_wawan >= laptop_oman:
  print("Beli laptop Oman")
else:
  print("Menabung dulu sampai uangnya cukup!")
```

**MINI EXERCISE**

Anda makan di sebuah restoran. Anda ingin memesan menu sesuai dengan budget anda.

*   Jika uang anda lebih dari sama dengan 100.000 maka anda memakan steak.

*   Jika uang anda lebih dari sama dengan 50.000 maka anda memakan spaghetti.
*   Jika uang anda kurang dari 50.000 maka anda memakan nasi goreng.

Budget uang anda sebesar 60.000 maka lengkapilah program branching dibawah ini:

```python
# Lengkapilah program branching dibawah ini agar bisa makan spaghetti

uang = 60_000

if uang >= 100_000:
  print("Makan steak")
elif uang >= 50_000:
  print("Makan spaghetti")
elif uang <= 50_000:
  print("Makan nasi goreng")
else:
  print("Puasa dulu")
```
