# **Pertemuan #6 - Looping**

**LOOP CONTROL FLOW**

Menggunakan perulangan di Python membuat kita bisa menyelesaikan tugas dengan cara yang efisien. Namun, mungkin muncul kondisi di mana kita ingin keluar dari kondisi perulangan, atau melewatkan sebuah perulangan bahkan jika hanya mengabaikan kondisi itu. Hal ini dapat dilakukan dengan Loop Control Flow. Loop Control Flow akan mengubah eksekusi dari urutan normalnya sesuai yang kita inginkan.

**BREAK, CONTINUE, PASS**

Apabila ada suatu kondisi dan mengharuskan kita untuk **menghentikan** loop atau perulangan secara langsung saat menemukan kondisi tersebut kita bisa menggunakan `break`.

`break` akan memaksa jalannya sebuah program untuk berhenti jika menemui kondisi tertentu.

Flowchart `break` bisa dilihat pada gambar dibawah:

dslfjalsd.JPG

**EXAMPLE**

```python
# Coba jalankan code di bawah
# Kita akan mencari element Malang pada list di bawah
# Dan jika sudah ditemukan, makan akan diakhiri proses loop tersebut

list_kota = ['Surabaya', 'Malang', 'Bandung']

for kota in list_kota:
    if kota == "Malang":
        break
    print(f'Hallo {kota}')
```

```python
kalimat = 'Doni suka makan ayam bakar dan sate'

# Coba hentikan perulangan di bawah ketika menemukan kata 'ayam' pada variabel kalimat

for kata in kalimat.split():
  if kata == "ayam":
    break
  print(kata, end=" ")
```

`continue` juga merupakan **Loop Control Flow** seperti `break`. Jika `break` memaksa iterasi untuk berhenti maka `continue` akan memaksa untuk **eksekusi iterasi berikutnya**.

Ketika pernyataan `continue` dieksekusi, kode di dalam loop setelah pernyataan `continue` akan **dilewati** dan iterasi berikutnya akan dimulai.

Flowchat `continue` bisa dilihat pada kode dibawah:

alsdfjlafs.JPG

```python
# Coba jalankan code dibawah ini
list_kota = ['Surabaya', 'Malang', 'Bandung']

for kota in list_kota:
    if kota == "Malang":
      continue
    print(f'Hallo {kota}')
```

```python
# Coba buatlah perulangan untuk menampilkan angka 1 - 10
# Namun kelipatan tiga tidak dimasukkan

for angka in range(1, 11):
  if angka % 3 == 0:
    continue
  print(angka, end=" ")
  angka += 1
```

Berbeda dengan `break` dan `continue` yang akan membuat perbedaan pada proses eksekusi program, `pass` **tidak melakukan apapun**. `pass` dalam Python digunakan ketika pernyataan diperlukan secara sintaksis tetapi kita tidak ingin perintah atau kode apa pun untuk dieksekusi.

`pass` bisa juga digunakan sebagai penanda bahwa blok kode tersebut telah dijalankan.

Flowchart `pass` bisa dilihat pada gambar dibawah:

jdlafjsld.JPG

```python
list_kota = ['Surabaya', 'Malang', 'Bandung', 'Jakarta', 'Yogyakarta', 'Semarang', 'Surakarta', 'Bogor']

for kota in list_kota:
    if kota == "Malang":
      pass
      print('Pass dieksekusi')
    print(kota)
```
